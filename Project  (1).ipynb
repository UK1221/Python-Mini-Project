```python
# Hotel Menu card
menu = {
    'Room Charges': {
        'Single Room': 100,
        'Double Room': 150,
        'Suite Room': 250
    },
    'Food Menu': {
        'Breakfast': 15,
        'Lunch': 25,
        'Dinner': 35
    },
    'Services': {
        'Wi-Fi': 10,
        'Laundry': 15,
        'Spa': 50
    }
}
invoice_number = 12345
date = "2025-01-13"
customer_name = "Utkarsh Khandare"
tax_rate = 10  
discount_rate = 5  
room_charge = 0
food_charge = 0
service_charge = 0
food_items = []
payment_mode = ""

while True:
    print("\nWelcome to the Hotel Menu")
    print("1. Room Charges")
    print("2. Food Menu")
    print("3. Services")
    print("4. Generate Invoice")
    print("5. Exit")
    choice = int(input("\nEnter your choice (1-5): "))
    
    if choice == 1:  # Room Charges
        print("\nAvailable Rooms:")
        for idx, room in enumerate(menu['Room Charges'], 1):
            print(f"{idx}. {room} - ${menu['Room Charges'][room]}")
        room_choice = int(input("\nSelect a room (1-3): "))
        if room_choice == 1:
            room_charge = menu['Room Charges']['Single Room']
        elif room_choice == 2:
            room_charge = menu['Room Charges']['Double Room']
        elif room_choice == 3:
            room_charge = menu['Room Charges']['Suite Room']
        else:
            print("Invalid choice!")
    
    elif choice == 2:  # Food Menu
        print("\nAvailable Food Items:")
        for idx, food in enumerate(menu['Food Menu'], 1):
            print(f"{idx}. {food} - ${menu['Food Menu'][food]}")
        food_choice = int(input("\nSelect food (1-3): "))
        if food_choice == 1:
            food_charge += menu['Food Menu']['Breakfast']
            food_items.append("Breakfast")
        elif food_choice == 2:
            food_charge += menu['Food Menu']['Lunch']
            food_items.append("Lunch")
        elif food_choice == 3:
            food_charge += menu['Food Menu']['Dinner']
            food_items.append("Dinner")
        else:
            print("Invalid choice!")
    
    elif choice == 3:  # Services
        print("\nAvailable Services:")
        for idx, service in enumerate(menu['Services'], 1):
            print(f"{idx}. {service} - ${menu['Services'][service]}")
        service_choice = int(input("\nSelect service (1-3): "))
        if service_choice == 1:
            service_charge = menu['Services']['Wi-Fi']
        elif service_choice == 2:
            service_charge = menu['Services']['Laundry']
        elif service_choice == 3:
            service_charge = menu['Services']['Spa']
        else:
            print("Invalid choice!")
    
    elif choice == 4:  # Generate Invoice
        print("\nSelect Payment Mode:")
        print("1. Online")
        print("2. Offline")
        payment_choice = int(input("Enter your choice (1-2): "))

        if payment_choice == 1:
            payment_mode = "Online"
        elif payment_choice == 2:
            payment_mode = "Offline"
        else:
            payment_mode = "Not Selected"
            print("Invalid payment mode selected!")
        total_amount = room_charge + food_charge + service_charge
        subtotal = total_amount
        tax_amount = subtotal * tax_rate / 100
        discount_amount = subtotal * discount_rate / 100
        total_after_discount = subtotal - discount_amount
        final_total = total_after_discount + tax_amount
        print("\n===================================")
        print(f"INVOICE # {invoice_number}")
        print(f"Date: {date}")
        print(f"Customer: {customer_name}")
        print(f"Payment Mode: {payment_mode}")
        print("-----------------------------------")
        print("Description          Price")
        print("-----------------------------------")
        if room_charge > 0:
            print(f"Room Charge         ${room_charge}")
        if food_charge > 0:
            print(f"Food Charges        ${food_charge}")
            print("Food Items Ordered: " + ", ".join(food_items))
        if service_charge > 0:
            print(f"Service Charges     ${service_charge}")

        print("-----------------------------------")
        print(f"Subtotal:           ${subtotal}")
        print(f"Discount ({discount_rate}%): -${discount_amount}")
        print(f"Tax ({tax_rate}%):  ${tax_amount}")
        print(f"Total Amount:       ${final_total}")
        print("===================================")
    
    elif choice == 5:  # Exit
        print("Thank you for visiting!")
        break
    
    else:
        print("Invalid choice! Please select between 1 to 5.")
```
